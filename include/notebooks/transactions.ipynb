```python
import pandas as pd
```

```python
df = pd.read_csv(r'..\include\data\transactions.csv')
df.infer_objects()

df['Date'] = pd.to_datetime(df['Date'])
print(df.head())
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
total_rows = len(df)
null_percent = df.isnull().sum() / total_rows
null_cols = null_percent[null_percent > 0.4].index
df_cleaned = df.drop(columns=null_cols)
df_cleaned
```

```python
df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).fillna(0)
df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).fillna('unknown')
```
